## polarity score 계산 함수

```
def calc_polarity(scores):
    eps = 1e-6
    pos_score = [1 for s in scores if s > 0]
    neg_score = [-1 for s in scores if s < 0]

    s_pos = sum(pos_score)
    s_neg = sum(neg_score)

    s_pol = ((s_pos + s_neg) * 1.0 /
             ((s_pos - s_neg) + eps))

    return s_pol
```

## tone 계산

```
%%time

mpko_mkt = MPKO(kind=0)
mpko_lex = MPKO(kind=1)

scores = []
for row in mpd_text_df.to_records(index=False):
    title, time, section, sid, sentence = row
    tokens = mpko_lex.tokenize(sentence)
    tone_lex = mpko_lex.get_score(tokens)['Polarity']
    tokens = mpko_mkt.tokenize(sentence)
    tone_mkt = mpko_mkt.get_score(tokens)['Polarity']
    score = (title, time, section, sid, tone_mkt, tone_lex)
    scores.append(score)
  
score_df = pd.DataFrame(scores, columns=['title', 'time', 'section', 'sid', 'tone_mkt', 'tone_lex'])

key_cols = ['time']
tone_cols = ['tone_mkt', 'tone_lex']
mpd_tone_mpko_df = score_df.groupby(key_cols)[tone_cols].agg(lambda x: calc_polarity(x)).reset_index()
```

### 날짜에 따른 매파/비둘기파 분류 dict 저장

```
from datetime import date
from dateutil.relativedelta import relativedelta

def check_hawk_dove(start_time):
    end_time = start_time + relativedelta(months=+1)

    start_callrate = callrate_df[callrate_df['time'] == start_time]['call_rate'].values[0]
    end_callrate = callrate_df[callrate_df['time'] == end_time]['call_rate'].values[0]

    change = ((end_callrate / start_callrate) - 1) * 100

    # 각주 25) 의미 없는 변화는 제외하기 위해 +/-3bp를 threshold로 함, 1bp = 0.01%
    if change >= 0.03:
        return 1    # hawkish
    elif change <= -0.03:
        return -1    # dovish
    else:
        return 0    # no significant change
```

```
callrate_hawk_dove_dict = {}

for start_time in callrate_df['time']:
    end_time = start_time + relativedelta(months=+1)

    if end_time <= max(callrate_df['time']):
        callrate_hawk_dove_dict[start_time] = check_hawk_dove(start_time)

callrate_hawk_dove_dict
```

### 전체 텍스트 데이터에 페이지 넘버 제거, KoSpacing 기준 띄어쓰기 적용(오래 걸림)

```
import re

from pykospacing import spacing

def clean_text(text):
    # 페이지 넘버 제거
    p = re.compile('-\s\d+\s-')
    m = p.findall(text)
    for item in m:
        text = text.replace(item, '')

    # 공백 제거
    text = text.replace(' ', '')

    # KoSpacing 띄어쓰기 적용
    text_clean = spacing(text)

    return text_clean
```

### 문장별 1- to 5-gram 생성

```
def generate_ngrams(pos_list, n):   
    # Use the zip function to help us generate n-grams
    # Concatentate the tokens into ngrams and return
    ngrams = list(zip(*[pos_list[i:] for i in range(n)]))

    return ngrams
```

```
from collections import defaultdict 

ngrams_tmp_dict = defaultdict(list)

def get_sent_ngrams(col):
    ngrams_tmp_dict[1].append(list(zip(col)))    # n=1
    for i in range(2, 6):
        ngrams = generate_ngrams(col, n=i)
        ngrams_tmp_dict[i].append(ngrams)
```
