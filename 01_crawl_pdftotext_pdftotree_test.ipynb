# scrapy로 크롤링한 금통위 pdf url로 데이터 읽어보기

* ```/content/drive/My Drive/2020NLP_Crawl/bok/bok_pdf_urls.tsv```

````
import pdftotext, pdftotree
````

````
import csv

data_dir = '/content/drive/My Drive/2020NLP_Crawl/'
url_file = data_dir + 'bok/bok_pdf_urls.tsv'

with open(url_file, 'r') as f:
    reader = csv.reader(f, delimiter='\t')
    minute_urls = {r[0]: r[1] for r in reader}
````

````
minute_urls
````

````
import os
import requests


def download(url: str, name: str, dest_folder: str):
    if not os.path.exists(dest_folder):
        os.makedirs(dest_folder)  # create folder if it does not exist

    # filename = url.split('/')[-1].replace(" ", "_")  # be careful with file names
    file_path = os.path.join(dest_folder, name)

    r = requests.get(url, stream=True)
    if r.ok:
        print("saving to", os.path.abspath(file_path))
        with open(file_path, 'wb') as f:
            for chunk in r.iter_content(chunk_size=1024 * 8):
                if chunk:
                    f.write(chunk)
                    f.flush()
                    os.fsync(f.fileno())
    else:  # HTTP status code 4XX/5XX
        print("Download failed: status code {}\n{}".format(r.status_code, r.text))
````

````
for name in sorted(minute_urls.keys():
    download(minute_urls[name], name + '.pdf', dest_folder=data_dir + 'bok/pdfs')
````

````
for minute in sorted(minute_urls.keys()):
    pdf_file = data_dir + 'bok/pdfs/' + minute + '.pdf'

    with open(pdf_file, 'rb') as f:
        pdf = pdftotext.PDF(f)
        # minute_data[minute] = pdf        
        minute_data[minute] = "\n\n".join(pdf)
````
